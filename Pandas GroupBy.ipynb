```python
def foo_new(group):
    return group['Series_Title'].str.startswith('A').sum()
```

```python
# Ranking of each movie in the group based on IMDB Rating

def rank_movie(group):
    group['Rank'] = group['IMDB_Rating'].rank(ascending = False)
    return group   
```

```python
genre.apply(rank_movie)
```

```python
# find normalized rating for each movie

def normal(group):
    group['norm_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min())/(group['IMDB_Rating'].max()-group['IMDB_Rating'].min()) 
    return group
```

```python
genre.apply(normal)
```

```python
# find the top 10 batsman in term of runs


ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
ipl.groupby('batsman').agg({'batsman_runs':'sum'}).sort_values('batsman_runs',ascending=False).head(10)
```

```python
# batsman with most number of sixes

six = ipl[ipl.batsman_runs == 6]

six.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1)
```

```python
six.batsman.value_counts().head(1)
```

```python
# find batsman with most number of 4s and 6s in last 5 overs

temp_df = ipl[ipl.over>15]

temp_df = temp_df[(temp_df.batsman_runs==4)|(temp_df.batsman_runs==6)]
```

```python
temp_df.batsman.value_counts().head(1)
```

```python
temp_df.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1)
```

```python
# find v kholi record against all the teams

temp_ipl_df = ipl[ipl['batsman'] == 'V Kohli']
```

```python
temp_ipl_df.groupby('bowling_team').agg({'batsman_runs':'sum'}).sort_values(by='batsman_runs',ascending = False).reset_index()
```

```python
# create a function that will return the highest score of any batsman


def highest(batsman):
    temp_df_batsman = ipl[ipl.batsman == batsman]
    return temp_df_batsman.groupby('match_id').agg({'batsman_runs':'sum'}).sort_values('batsman_runs',ascending=False).head(1).values[0]
```

```python
highest("CH Gayle")
```
