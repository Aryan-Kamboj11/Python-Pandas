```python
regs = pd.concat([nov,dec]).reset_index(drop=True)
```

```python
pd.merge(students,regs,'inner','student_id')
```

```python
courses.merge(regs,'left','course_id')
```

```python
# find total revenue generated by the company.

regs.merge(courses,'inner','course_id')['price'].sum()
```

```python
regs.merge(courses,'inner','course_id').groupby('course_name').agg({'price':'sum'}).reset_index()
```

```python
# find month by month revenue
multi_index_df.merge(courses,'inner','course_id').groupby('level_0').agg({'price':'sum'}).reset_index()
```

```python
# print the registration table 

# columns - > student_name, course , price
regs.merge(students,on='student_id').merge(courses,on='course_id')[['name','course_name','price']]
```

```python
# plot a bar chart for revenue/courses
regs.merge(courses,'right','course_id').groupby('course_name').agg({'price':'sum'}).plot(kind='bar')
```

```python
# find the students who enrolled in both the months

common_stu_id=np.intersect1d(nov['student_id'],dec['student_id'])
```

```python
students[students['student_id'].isin(common_stu_id)]
```

```python
# find courses that got no enrollment

course_id_list=np.setdiff1d(courses['course_id'],regs['course_id'])
```

```python
courses[courses['course_id'].isin(course_id_list)]
```

```python
# find student who did not enroll in any course
student_id_list=np.setdiff1d(students['student_id'],regs['student_id'])
```

```python
students[students['student_id'].isin(student_id_list)]
```

```python
#Print student name - >  print partner name of all enrolled students

#self join 

students.merge(students,'inner',left_on='partner',right_on='student_id')[['name_x','name_y']]
```

```python
# find top 3 students who did most enrollments

regs.merge(students,on='student_id').groupby(['student_id','name'])['name'].count().sort_values(ascending=False).head(3)
```

```python
# 3 students who spent most money

regs.merge(students,on='student_id').merge(courses,on='course_id').groupby(['student_id','name'])['price'].sum().sort_values(ascending=False).head(3)
```

```python
# find top 3 matches with highest sixes/match ratio
# find orange cap holder of all season

temp_df=matches.merge(deliveries,left_on='id',right_on='match_id')
```

```python
temp_df=temp_df[temp_df['total_runs']==6]
```

```python
num_sixes = temp_df.groupby('venue')['venue'].count().sort_values(ascending=False)
```

```python
num_matches = matches['venue'].value_counts()
```

```python
(num_sixes/num_matches).sort_values(ascending=False).head(3)
```

```python
# find orange cap holder of all season

matches.merge(deliveries,left_on = 'id',right_on = 'match_id').groupby(['season','batsman'])['batsman_runs'].sum().reset_index().sort_values('batsman_runs',ascending=False).drop_duplicates(subset=['season'],keep='first').sort_values('season')
```
